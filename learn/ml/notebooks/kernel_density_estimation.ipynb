# Kernel Density Estimation

```python
%matplotlib inline

import numpy as np
import sklearn.neighbors
from matplotlib import pyplot as plt

from labm8.py import viz
```

## Data

```python
sample = np.hstack([
    np.random.normal(loc=10, scale=5, size=300),
    np.random.normal(loc=30, scale=5, size=700),
])
sample = sample.reshape((len(sample), 1))
print(sample.shape)

plt.hist(sample, bins=50, density=True)
viz.Finalize()
```

## Kernel Density Estimation

```python
values = np.arange(50)
values = values.reshape((len(values), 1))

plt.hist(sample, bins=50, density=True)

for bandwidth in [1, 2, 3, 5, 10]:
    model = sklearn.neighbors.KernelDensity(bandwidth=bandwidth, kernel='gaussian')
    model.fit(sample)
    probabilities = np.exp(model.score_samples(values))
    plt.plot(values, probabilities, label=f"bandwidth={bandwidth}")

plt.legend()
viz.Finalize()
```
