## Load and use model

```python
from third_party.py.tensorflow import tf

def load_graph(frozen_graph_filename):
    # We load the protobuf file from the disk and parse it to retrieve the 
    # unserialized graph_def
    with tf.gfile.GFile(frozen_graph_filename, "rb") as f:
        graph_def = tf.GraphDef()
        graph_def.ParseFromString(f.read())

    # Then, we import the graph_def into a new Graph and returns it 
    with tf.Graph().as_default() as graph:
        # The name var will prefix every op/nodes in your graph
        # Since we load everything in a new graph, this is not needed
        tf.import_graph_def(graph_def, name="prefix")
    return graph
```

```python
graph = load_graph('/tmp/export/frozen.pb')
input_data = graph.get_tensor_by_name('prefix/Placeholder:0')
lengths = graph.get_tensor_by_name('prefix/Placeholder_2:0')
seed_length = graph.get_tensor_by_name('prefix/Variable_1:0')
output = graph.get_tensor_by_name('prefix/rnnlm_1/transpose_1:0')
```

```python
import numpy as np

atomizer = model.atomizer

batch_size = 64
sequence_length = 1024

encoded_start_text = atomizer.AtomizeString('kernel void ')
inference_indices = np.tile(encoded_start_text, [batch_size, 1])

sample = inference_indices.copy()

with tf.compat.v1.Session(graph=graph) as s:    
    length = inference_indices.shape[1]

    for _ in range(10):
        expanded_indices = np.zeros((batch_size, sequence_length))
        expanded_indices[:, :length] = inference_indices
        synthesized_lengths = np.full([sampler.batch_size], length)    
        feed = {
            input_data: expanded_indices,
            lengths: synthesized_lengths,
            seed_length: length,
        }

        generated, = s.run([output], feed)
        inference_indices = generated[:, -1].reshape((batch_size, 1))
        if length > 1:
            generated = generated[:, length-1:]

        sample = np.concatenate((sample, generated), axis=1)
        
```
